```python
def fillhole(X,Y):
    Y=Y.interpolate()
    
    X.reset_index(inplace=True)
    Y.reset_index(inplace=True)
    Y_return=Y.value
    X_return=X.value
    x_column=list(X.columns).index('value')
    y_column=list(Y.columns).index('value')
    ret=pd.concat([X_return,Y_return],axis=1)
    ret.columns=['X_ret','Y_ret']
    ret_na=ret.dropna()
    Y_ret=ret_na.Y_ret
    X_ret=ret_na.X_ret
    Y_ret = sm.add_constant(Y_ret, prepend=False)
    model = sm.OLS(X_ret,Y_ret)
    res=model.fit()
    Y_return=sm.add_constant(Y_return, prepend=False)
    for i in range(X.shape[0]):
        if pd.isnull(X.iloc[i,x_column]):
            X.iloc[i,x_column]=Y_return.iloc[i,:].dot(np.array(res.params))
    X.set_index('index',inplace=True)
    return [X,res]
```

```python
def r_sqaure(y_true,y_fit):
    SSR=(y_fit-st.mean(y_true)).dot(y_fit-st.mean(y_true))
    TSR=(y_true-st.mean(y_true)).dot(y_true-st.mean(y_true))
    return SSR/TSR
```
